```python
from pathlib import Path
import pandas as pd
import numpy as np
import datetime as dt
# import matplotlib.pyplot as plt
```

```python
p = Path("C:/Users/k/Desktop/GitHub/GPS_transport/")

aug_data = pd.read_hdf(p / "processed_data" / "aug_all.h5", key="aug")
segments = pd.read_csv(p / "input_data" / "python" / "segments.csv")
markers = pd.read_csv(p / "input_data" / "python" / "markers.csv")
```

```python
# disregarding dir = 2 data
aug_data = aug_data.loc[aug_data["dir"] == 1]
```

```python
time_bin = "30T"
aug_data = aug_data.groupby("segment_id").apply(
                                            lambda x: x.set_index("aug_time")["speed"].resample("30T").mean()
                                            )
```

```python
aug_data = aug_data.reset_index()
aug_data
```

```python
df_list = [pd.pivot_table(group_data, values="speed", columns="segment_id", index="aug_time") for _, group_data in aug_data.groupby("segment_id")]
```

```python
segment_speed = pd.concat(df_list, axis=1)
```
